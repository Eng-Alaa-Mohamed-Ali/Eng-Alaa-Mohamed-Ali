## اسم المشروع:
**Breast Cancer Classification Model**

## الهدف:
بناء نموذج ذكاء اصطناعي قادر على تصنيف الأورام (خبيثة أو حميدة) بناءً على الخصائص الطبية المتاحة في مجموعة بيانات Sklearn.

## أبرز ما تم في المشروع:

*   **استكشاف البيانات**: التعامل مع 30 خاصية طبية مختلفة (Features).
*   **هندسة البيانات**: تقسيم البيانات بنسبة 33% للاختبار لضمان دقة النتائج وعدم الانحياز.
*   **الخوارزمية**: اختيار Gaussian Naive Bayes كنموذج إحصائي لبناء التوقعات.
*   **التقييم**: قياس كفاءة النموذج باستخدام Accuracy Score للتأكد من مدى موثوقية التشخيص.

```python
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

```

```python
#تحميل البيانات
data=load_breast_cancer()
```

```python
#تنظيم البيانات
lable_names=data['target_names']
labels=data['target']
feature_names=data['feature_names']
features=data['data']
```

```python
#الاطلاع على البيانات
print(lable_names)
print(labels[0])
print(feature_names[0])
print(features[0])
```

```python
#هتقسيم البيانات
train , test, train_labels, test_labels=train_test_split(features,labels,test_size=0.33,random_state=42)
```

```python
#تهيئة المصنّف
gnb=GaussianNB()
model=gnb.fit(train,train_labels)
predict=model.predict(test)
print(predict)
```

```python
#تدريب المصنّف
model=gnb.fit(train,train_labels)
```

```python
#بناء التوقعات
predict=model.predict(test)
print(predict)
```

```python
#تقييم الدقة
print(accuracy_score(test_labels,predict))
```
